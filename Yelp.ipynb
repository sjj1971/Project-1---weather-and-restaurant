```python

for index, row in checkin_data.iterrows():
    business_id = row['business_id']
    result = requests.get(yelp_base_url + business_id, headers = yelp_key).json()
    test_list = ""
    try:
        checkin_data.loc[index, 'Name'] = result['name']
        checkin_data.loc[index, 'City'] = result['location']['city']
        checkin_data.loc[index, 'State'] = result['location']['state']
        #checkin_data.loc[index, 'Category'] = result['categories'][0]['title']
        for i in range(len(result["categories"])):
            test_list = test_list + " " + str(result["categories"][i]["alias"])
        checkin_data.loc[index, "Category"] = test_list
        checkin_data.loc[index, 'Rating'] = result['rating']
        checkin_data.loc[index, 'Review Count'] = result['review_count']
    except:
        print("Restaurant does not have enough data.")
```

```python
checkin_data = checkin_data.dropna() #removing Null Values
#cities las vegas, charlotte, toronto
#checkin_data['City']
project_data = checkin_data.loc[((checkin_data['City'] == 'Las Vegas') | (checkin_data['City'] == 'Charlotte') | (checkin_data['City'] == 'Toronto'))]
project_data = project_data.rename(columns = {'business_id': 'Business ID',
                                             'total_checkin': 'Total Check-Ins',
                                             '01': '2019-01',
                                             '02': '2019-02',
                                             '03': '2019-03',
                                             '04': '2019-04',
                                             '05': '2019-05',
                                             '06': '2019-06',
                                             '07': '2019-07',
                                             '08': '2019-08',
                                             '09': '2019-09',
                                             '10': '2019-10',
                                             '11': '2019-11',
                                             '12': '2019-12'})
checkin_data.to_csv("YelpData/unfiltered_data.csv")
project_data.to_csv("YelpData/filtered_data.csv")
project_data
```
