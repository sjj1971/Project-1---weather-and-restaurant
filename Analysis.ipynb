```python
import pandas as pd
```

```python
city_cuisines = pd.read_csv("YelpData/CityCuisines.csv").rename(columns={'Unnamed: 0': 'city'})
city_cuisines = city_cuisines.drop(['seafood', 'cafe'], axis = 1)
city_cuisines
```

```python
filtered_data = pd.read_csv("YelpData/filtered_data.csv").drop('Unnamed: 0', axis = 1)
#filtered_data.head()
```

```python
ch_cuisines = city_cuisines.loc[city_cuisines['city'] == 'Charlotte']
#filtered_data.loc[filtered_data['Business ID'] == LV_cuisines['japanese'][0]]

cuisines = pd.DataFrame()
cuisine_list = city_cuisines.columns[1:]
```

```python
for index, row in city_cuisines.iterrows():
    city = row['city']
    temp_city_df = city_cuisines.loc[city_cuisines['city'] == city]
    for cuisine in cuisine_list:
        try:
            cuisines = pd.concat([cuisines, filtered_data.loc[filtered_data['Business ID'] == temp_city_df[cuisine][index]]])
        except:
            print(f"That city does not have any {cuisine} cuisines!")
cuisines
```

```python
cuisines_by_city = cuisines.set_index('City')
cuisines_by_city.to_csv("FinalData/cuisines_by_city.csv", index = True)
cuisines.to_csv("FinalData/cuisines.csv")
```
